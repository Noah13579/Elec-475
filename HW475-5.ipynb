```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA, FastICA

np.random.seed(0)
t = np.arange(0, 25, 0.001)

S1 = np.cos(t)
S2 = np.exp(-t) - 5*np.exp(-t/5)

S = np.c_[S1, S2]

X1 = 0.8*S1 + 0.5*S2
X2 = 0.3*S1 - 0.4*S2
X = np.c_[X1, X2]

ica = FastICA(n_components = 2, whiten = 'arbitrary-variance')
S_ = ica.fit_transform(X)

plt.plot(t, S)
plt.title("Source Signal Plot")
plt.xlabel("Time")
plt.ylabel("Amplitudes")
plt.show()
```

```python
plt.plot(t, X)
plt.title("Observation Signal Plot")
plt.xlabel("Time")
plt.ylabel("Amplitudes")
plt.show()
```

```python
plt.plot(t, S_)
plt.title("Recovered Signal Plot")
plt.xlabel("Time")
plt.ylabel("Amplitudes")
plt.show()
```
